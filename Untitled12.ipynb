```python
import pandas as pd
import quandl
```

```python
df = quandl.get('WIKI/GOOGL')
```

```python
df.head()
```

```python
df = df[['Adj. Open', 'Adj. High', 'Adj. Low', 'Adj. Close', 'Adj. Volume']]  #DataFrame with columns of interest
```

```python
df.head()
```

```python
df['HL_PCT'] = (df['Adj. High'] - df['Adj. Close']) / df['Adj. Close'] * 100.0  #High and Low percent change
```

```python
df['PCT_change'] = (df['Adj. Close'] - df['Adj. Open']) / df['Adj. Open'] * 100.0  #Percent change
```

```python
df = df[['Adj. Close', 'HL_PCT', 'PCT_change', 'Adj. Volume']]
```

```python
df.head()
```

```python
df.info()
```

```python
forecast_col = 'Adj. Close'
```

```python
df.isnull( ).sum()
```

```python
df.duplicated().sum()
```

```python
import math
```

```python
forecast_out = int(math.ceil(0.01* len(df)))   #Number of days out using 10% of the DataFrame (using data from 10 days ago)
```

```python
df['label'] = df[forecast_col].shift(-forecast_out)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
import numpy as np
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
 X = np.array(df.drop(['label'], axis=1))  #Features
 y = np.array(df['label'])  #Labels
```

```python
X = preprocessing.scale(X)    #Scaling features
```

```python
X
```

```python
df.isnull().sum()
```

```python
len(X), len(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred= model.predict(X_test)
```

```python
print ("The model intercept is :", model.intercept_)
print ('The model coefficient is:', model.coef_)
```

```python
compare= pd.DataFrame({'Actual': y_test, 'predict': y_pred})
compare. head()
```

```python
from sklearn import metrics
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Error evaluation
print ('Mean Absolute Error:', metrics.mean_absolute_error (y_test, y_pred))
print ('Mean Squard Error:', metrics.mean_squared_error (y_test, y_pred))
print ('Root Mean Squard Error:', np.sqrt (metrics.mean_squared_error (y_test, y_pred)))
print ('R2 Score:', metrics.r2_score(y_test, y_pred))
```

```python
sns.regplot(x=y_pred, y=y_test)
plt.xlabel ('Predicted')
plt.ylabel ('Actual')
plt.title('Actual VS predicted price')
```
